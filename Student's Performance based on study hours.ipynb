```python
# importing required libraries 1
import pandas as pd
import numpy as np
```

```python
# reading file 
df = pd.read_csv("score.csv")
df.head()
```

```python
# importing necessary librry for visualization
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fig,ax= plt.subplots(figsize = (10,5))
ax.scatter(df.Hours,df.Scores)
plt.xlabel("Study Hours")
plt.ylabel("Marks Obtained")
plt.title("Student's Performance based on study hours")

plt.show()
```

```python
plt.subplots(figsize = (10,5))
sns.regplot(data=df,x="Hours",y="Scores")

plt.show()
```

```python
# importing necessary library for training, testing and predicting
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
#  Preparing the data 
x= df.iloc[:,0:1]
y =df.iloc[:,1:]

# Splitting the data into training and testing datasets
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2)
```

```python
# Checking data and its shape before working in it
print(x_train.shape, y_train.shape)
print(x_test.shape, y_test.shape)
```

```python
x_train.head()
```

```python
# Selecting the model and fitting data in it

model = LinearRegression()
model.fit(x_train, y_train)
```

```python
# Checking the accuracy of prediction 
model.score(x_test,y_test)
```

```python
# Checking model by giving new unseen data to predict
x_new = [[0.5]]
scr = model.predict(x_new)
```

```python
# printing out the result
print("If a student study for approx "+str(x_new)[2:-2]+" hours then the student can score around "+ str(scr)[2:-2]+ " marks.")
```
